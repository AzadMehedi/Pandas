```python
import pandas as pd
import numpy as np
```

# Crating Dataframe
table == dataframe

any structure that has row and columns is a dataframe

Series just got one column

```python
# using list
# need 2d array to make a dataframe
student_data = [
    [100, 80, 10],
    [90, 70, 7],
    [120, 100, 14],
    [80, 50, 2]
]

student_data = pd.DataFrame(student_data, columns=['iq','marks','package'])
student_data
```

```python
# using dictionary
student_dict = {
    'iq':[100, 90, 120, 80],
    'marks':[80, 70, 100, 50],
    'package':[10, 7, 14, 2]
}
pd.DataFrame(student_dict)
```

```python
# using read_csv
movies = pd.read_csv('/content/movies.csv')
movies
```

```python
# dtype    -> shows datatypes of all columns
movies.dtypes
ipl.dtypes
```

```python
# index    -> shows number of indices
movies.index
ipl.index
```

```python
# column      -> shows all the columns
movies.columns
ipl.columns
```

```python
student_data.columns
```

```python
# values      -> shows all the values in 2d array (dataframe is consist of 2d array)
student_data.values
ipl.values
```

```python
# head()      -> see first 5(default) data
movies.head()
```

```python
# tail()        -> see last 5(default) data 
movies.tail(2)
```

```python
# sample()      ->  randomly select
ipl.sample(7)
```

```python
# isnull()        -> to find is amy missing values in dataframe
movies.isnull().sum()
```

```python
# duplicated() -> shows any duplicated values in the columns
movies.duplicated().sum()
```
