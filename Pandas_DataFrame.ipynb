- **iloc** - searches using index positions
- **loc** - searches using index labels 

```python
# single row
movies.iloc[0]
```

```python
type(movies.iloc[0])     # index position using iloc
```

```python
movies.iloc[0:5]      # 5 is not included(index 0-4)
```

```python
movies.iloc[5:16:2]
```

```python
movies.iloc[5:]
```

```python
# loc      -> works label wise

students.loc['Evan']
```

```python
students.loc['Mehedi':'Azad']
```

```python
students.loc['Mehedi':'Azad':2]
```

`basic Difference between iloc and loc`
- `iloc` : last value is not included
- `loc` : last value is included
