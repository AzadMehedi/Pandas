```python
# pd.concat
# df.concat
# ignore_index
# df.append
# mullitindex -> fetch using iloc
# concat dataframes horizontally
```

```python
 # pd.concat      -> Vertical concat

pd.concat([nov, dec])
```

```python
# ignore_index
regs = pd.concat([nov, dec], ignore_index=True)
regs
```

```python
# df.append
nov.append(dec, ignore_index=True)
```

```python
# mullitindex -> fetch using iloc
multi = pd.concat([nov, dec], keys=['Nov','Dec'])     # prividing index
multi
```
